```python
###Setting The Problem Parameters
```

```python
N = 7
MAX_WEIGHT = 15
objects = [(10,2),(5,3),(15,5),(7,7),(6,1),(18,4),(3,1)]
#object = None
```

```python
###Item Class
```

```python
class Item:
    def __init__(self, profit, weight):
        self.profit = profit
        self.weight = weight
        self.profit_by_weight = profit/weight
        self.portion = 0.0
```

```python
###Item Getter Function
```

```python
def get_item(n, input_items=None, verbos=0):
    items = []
    if input_items==None:
        for i in range(n):
            print(f"Items {i+1}")
            item_profit = int(input("Please Enter the Profit of the item: "))
            item_weight = int(input("Please Enter the Weight of the item: "))
            items.append(Item(item_profit, item_weight))
    
    else:
        for item in input_items:
            items.append(Item(item[0],item[0]))
    if verbos:
        for item in items:
            print(f"Item {items.index(item)+1}: profit:{item.profit} weight:{item.weight}")
    return items
```

```python
###Highest Profit by Weight finder Function
```

```python
def highest_p_w_item(list_of_items):
    best_item = None
    for i in list_of_items:
        if best_item:
            if i.profit_by_weight > best_item.profit_by_weight and i.portion < 1:
                best_item = i
        else:
            if i.portion < 1:
                best_item = i

    return best_item 
```

```python
###Main
```

```python
if __name__ == "__main__":
    print(f"MAX WEIGHT is {MAX_WEIGHT}")
    print("*****************************************************************")
    items = get_item(N, input_items=None, verbos=0)
    bag = []
    OFS = MAX_WEIGHT
    while OFS>0:
        best_case = highest_p_w_item(items)
        if best_case:
            if best_case.weight <= OFS:
                best_case.portion = 1
                bag.append(best_case)
                OFS -= best_case
            else:
                best_case.portion = ((OFS * 100)/best_case.weight)/100
                bag.append(best_case)
                OFS = 0
        else:
            break

total_profit = 0
weights = ""
profits = ""

for i in bag:
    total_profit += i.portion * i.profit
    weights += f"{i.weight * i.portion}  + "
    profits += f"{i.profit * i.portion} + "
print("total progit: ", total_profit)
print(f"weights: {weights[:len(weights)-2]}")
print(f"profits: {profits[:len(profits)-2]}")
```
